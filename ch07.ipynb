#錯誤處理
>try:程式碼

>except:例外的時候處理動作

>else:當沒有進入例外(程式正確)的處理動作

>finally:不管正確例外，都會執行的動作

```python
try:
  a=10
  b=20
  if a<b:
    print(n)
except NameError:
  print("錯誤產生")
```

#指定錯誤類別處理

```python
a=10
b=20
try:
  if a<b:
    print(n)
except NameError as e:
  print("錯誤產生，錯誤原因:",e)
except ZeroDivisionError:
  print("除以0錯誤")
```

```python
try:
  a=10
  b=20
  if a<b:
    print("a<b")  #改變比較條件，讓錯誤不發生
except NameError:
  print("NameError錯誤產生")
except:
  print("其他錯誤產生")
else:
  print("沒有任何錯誤產生")
finally:
  print("不管程式正確錯誤都會執行的部分")
```
