#Tuple (元組)

tuple(元組)可視為不可改變的串列，tuple跟list很像，但是我們不能新增、刪除或更新tuple的元素。tuple的好處有：
1.  佔用較少空間
2.  可當作dictionary(字典)的鍵值
3.  可當作函式引數



```python
a_tuple=('python', 'js','SQL')
print(a_tuple)
b_list=['go','c#','vb']
b_tuple=tuple(b_list)
print(b_tuple)
print(b_tuple[0])
```

#字典(dictionary - dict)為帶有鍵值(key)的串列(list)

```python
langusges={}
a={0:'python', 1:'java', 2:'SQL'}
print(a[0])
print(a[1])
print(a[2])

b={"age":47,"gender":"M","name":"suhsien"}
print(b["age"])
print(b["name"])
print(b["gender"])

```

```python
langusges={'name':'python', 'version':"3.7"} #索引值(鍵值)可以自訂
print(langusges['name'])
print(langusges['version'])
print(langusges)
print(langusges.keys())
print(langusges.values())
print(langusges.items())
print('name' in langusges)
```
