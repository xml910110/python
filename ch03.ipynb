```python
score=[90,100,88,60,79]
print(score[0])
print(score[1])

for i in range(len(score)):  #len(score): score長度
  print(score[i])
```

```python
#score=[90,100,88,60,79]
print(score[0:3])  #取出score[0]-score[2]，不包含score[3]
print(score[:])   #取出所有的值
print(score[:4])  
print(score[2:])  #取出score[2]後面所有的值
print(score[0:5:2]) #間隔2 
print(score[::2])  
print(score[::-2])  #間隔-2，代表從列表最後往前取值，間隔2
print(score[4:0:-2])
```
